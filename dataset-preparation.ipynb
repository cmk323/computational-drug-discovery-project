Read the ChEMBL bioactivity data for Mycobacterium tuberculosis polyketide synthase Pks13 from the Pks13_bioactivity_data_2class_pIC50.csv file created in the Exploratory Data Analysis. This file contains the pIC50 values to be used for building a regression model.

```python
import pandas as pd
```

```python
df3 = pd.read_csv('Pks13_bioactivity_data_2class_pIC50.csv')
```

```python
selection = ['canonical_smiles','molecule_chembl_id']
df3_selection = df3[selection]
df3_selection.to_csv('molecule.smi', sep='\t', index=False, header=False)
```
